```python
import numpy as np
import pandas as pd 
```

## Gerando dados

```python
# Gerando um Numpy Array
npArray = np.random.rand(100000, 5)
npArray
```

```python
# Criando um Pandas DataFrame utilizando o array
pdDataFrame = pd.DataFrame(npArray, columns=['A', 'B', 'C', 'D', 'E'])
pdDataFrame
```

## Especificando linhas

```python
pdDataFrame[-5:]
```

## Visualizando os primeiros

```python
# Visualizando as primeiras linhas utilizando head
pdDataFrame.head(5)
```

```python
# Visualizando as primeiras linhas utilizando []
pdDataFrame[:5]
```

## Visualizando os últimos

```python
# Visualizando as últimas linhas utilizando tail
pdDataFrame.tail(5)
```

```python
# Visualizando as últimas linhas utilizando []
pdDataFrame[-5:]
```

## Aplicações

```python
# Visualizar após ordenações
pdDataFrame.sort_values(['A']).head()
```

```python
# Visualizar após ordenações
pdDataFrame.sort_values(['A'], ascending=False)[-5:]
```
